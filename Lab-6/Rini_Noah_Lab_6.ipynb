```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
```

```python
PATH = Service("/Users/noahrini/Documents/HARP 151/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

website = "https://www.binghamton-ny.gov/home"
driver.get(website)

menu = driver.find_element(By.XPATH, "//*[@id='dropdownrootitem3']/a")
submenu_departments = driver.find_element(By.XPATH, "//*[@id='dropdownrootitem3']/div/div/ul[1]/li/a")


actions = ActionChains(driver)
actions.move_to_element(menu)
time.sleep(5)
actions.move_to_element(submenu_departments)
time.sleep(5)
actions.click(submenu_departments)
actions.perform()


import csv
    
new_file = open("binghamton_jobs.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(new_file)
csv_writer.writerow(["job_title", "job_type", "apply_deadline", "salary"])



try:
    civil_service = WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id='widget_4_33_127']/ul/li[16]/a"))
    )
    
    civil_service.click()
    
    employment = WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id='leftNav_1038_0_145']/ul/li/ul/li[14]/ul/li/a"))
    )
    
    employment.click()
    
    table = WebDriverWait(driver, 30).until(
    EC.presence_of_element_located((By.TAG_NAME, "tbody"))
    )
    
    rows = table.find_elements(By.TAG_NAME, "tr")
        
    for i in rows:
        
        data = i.find_elements(By.TAG_NAME, "td")
        
        entry = []
        
        for i in data:
            
            print(i.text)
            entry.append(i.text)
            
            if len(entry) == 4:
                
                csv_writer.writerow(entry)
                
                entry.clear() 
        
    new_file.close()
            
                
                
finally:
    driver.quit()
```
