```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

PATH = Service("/Users/noahrini/Documents/HARP 151/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

website = "https://techwithtim.net"
driver.get(website)
print(driver.title)

search = driver.find_element(By.NAME, "s")
search.clear() # clears any text in search bar
search.send_keys("test")
search.send_keys(Keys.RETURN)

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

try:
    main = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "main"))
    )
    print(main.text)
    
    articles = main.find_elements(By.TAG_NAME, "article")
    
    for i in articles:
        header = i.find_element(By.CLASS_NAME, "entry-summary")
        print(header.text)
        
finally:
    driver.quit()
```

```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

PATH = Service("/Users/noahrini/Documents/HARP 151/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

website = "https://techwithtim.net"
driver.get(website) # ^ copying and pasting from above so I can keep everything together ^

# tutorial begins here:

link = driver.find_element(By.LINK_TEXT, "Python Programming") # storing the link we want to access
link.click() # executing click/opening above link


from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# want to then click on "Beginner Python Tutorials" :

# use try statement to allow time for the next link to appear/exist before trying to click it
try:
    # first, going to "Beginner Python Tutorials" link, so use By.LINK_TEXT:
    element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.LINK_TEXT, "Beginner Python Tutorials"))
    )
    element.click()
    
    # next, want to click on the "Get Started" button
    # inspect the HTML code to find element ID, then locate by using By.ID:
    element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "sow-button-19310003"))
    )
    element.click()
    
    driver.back() # goes back to previous page
    driver.back()
    driver.back() # use this three times to go all the way back to home page
    
    # driver.forward() : can also use this if needed, opposite of go back
    # element.clear() can be used to clear a text field if needed
    
except:
    driver.quit()
```

```python
PATH = Service("/Users/noahrini/Documents/HARP 151/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

website = "https://hoopshype.com/"
driver.get(website)

try:
    element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.LINK_TEXT, "SALARIES"))
    )
    element.click()
    
    element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.LINK_TEXT, "PLAYER SALARIES"))
    )
    element.click()
    
except:
    driver.quit()
    
import csv
    
new_file = open("nba_salaries.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(new_file)
csv_writer.writerow(["salary_rank", "player_name", "salary"])
    
try:
    
    table = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.TAG_NAME, "tbody"))
    )
    
    rows = table.find_elements(By.TAG_NAME, "tr")
    
    for i in rows:
        rank = i.find_element(By.CLASS_NAME, "rank")
        rank = rank.text
        
        name = i.find_element(By.CLASS_NAME, "name")
        name = name.text
        
        salary = i.find_element(By.CLASS_NAME, "hh-salaries-sorted")
        salary = salary.text
        
        csv_writer.writerow([rank, name, salary])
        
finally:
    driver.quit()
```
