```python
from tkinter import *
import json
import requests

root = Tk()
root.geometry("500x500")

t = Text(root, height = 5, width = 30, bg = "lightblue", font = "Georgia", padx = 100, pady = 100)
t.pack()

author_dict = {}

def get_quotes():

    author_dict.clear()
    
    random_quotes = requests.get(f"https://api.quotable.io/quotes/random?limit=5").json()

    for i in random_quotes:
    
        author = i['author']
        
        global slug
        slug = i['authorSlug']
        
        author_dict[author] = slug
        
get_quotes()

def random_authors():
        
    t.delete(0.0, END)
    
    author = value.get()
    
    quote_info = requests.get(f"https://api.quotable.io/random?author={slug}").json()
    
    quote = quote_info['content']
    
    t.insert(END, quote)
    
    get_quotes()
    
    


authors = author_dict.keys()
    
value = StringVar()
value.set("Select an Author")

dropdown = OptionMenu(root, value, *authors)
dropdown.pack()    
    
button = Button(root, text = "Get Quote", command = random_authors)
button.pack()


root.mainloop()
```

```python
from tkinter import *
import json
import requests


lat = 41.3645
lon = -74.1050
weather = requests.get(f"https://api.weather.gov/points/{lat},{lon}")
                      
json_file = weather.json()

forecast = json_file['properties']['forecast'] # accessing nested dictionary value

new_request = requests.get(forecast)
json = new_request.json()
json

forecasts = json['properties']['periods']
forecasts

days_of_week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]

days = []
temps = []
units = []

for i in forecasts:
    
    time = i.get('name')
    
    if time in days_of_week:
        
        days.append(time)
    
        temperature = i.get('temperature')
        temps.append(temperature)
        
        unit = i.get('temperatureUnit')
        units.append(unit)
    
        print(time)
        print(f"{temperature}{unit}")
        
    else:
        pass




root = Tk()
root.geometry("1150x600")

header = Label(root, text = "WEATHER IN WOODBURY, NY")
header.grid(row = 1, columnspan = 7)

n = len(days)

for i in range(0, n):
    
    text = Text(height=3, width=22, bg="lightblue")
    text.grid(row = 2, column = i+1)
    
    day = days[i] 
    text.insert(END, day)

for i in range(0, n):
    
    text2 = Text(height=3, width=22, bg="yellow")
    text2.grid(row = 3, column = i+1)
    
    temp = temps[i]
    temp_full = f"{temp}F"
    text2.insert(END, temp_full)
    
for i in range(0, n):
    
    description = Text(height=15, width=22, bg="yellow", relief = "groove")
    description.grid(row = 4, column = i+1)
    
    day = days[i]

    for i in forecasts:
    
        if i['name'] == day:
        
            detail = i['detailedForecast']
            description.insert(END, detail)
        
        else:
            pass
        
    
root.mainloop()
```
