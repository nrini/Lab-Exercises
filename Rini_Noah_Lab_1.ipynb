```python
# Problem 1

def longest_word(list):
    
    count_dict = {}
    count = []
    for i in list:
        n = len(i)
        count_dict[i] = n
        count.append(n)
        
    most = max(count)
    
    longest_list = []
    for i in count_dict:
        if count_dict[i] == most:
            longest_list.append(i)
            
    return longest_list
            


list = ["apple", "chestnut", "gargoyle", "pandas", "sheep", "raptor"]

longest_word(list)
```

```python
# Problem 4

# taking in student's personal information as the attributes

class StudentRecord:
    
    def __init__(self, last_name, first_name, id_num, year, gpa, major, grad_year, state, current_classes):
        self.last_name = last_name
        self.first_name = first_name
        self.id = id_num
        self.year = year
        self.gpa = gpa
        self.major = major
        self.grad_year = grad_year
        self.state = state
        self.current_classes = current_classes
        
# creating relevant methods for an academic advisor to use:

# the first will print an overview of the student's personal info
    
    def overview(self):
        
        return f"Student: {self.last_name}, {self.first_name}, ID #{self.id}, Standing: {self.year}, GPA: {self.gpa}, Major: {self.major}, Expected Graduation: {self.grad_year}, State of Residence: {self.state}."
    
# the second will calculate roughly how many semesters until they're graduating
# (assuming all students graduate in 4 years)
    
    def grad_check(self):
        
        sems_to_grad = (self.grad_year - 2023)*2
        return f"{self.first_name} will graduate in about {sems_to_grad} semesters."
    
# this method allows to add a class to the student's current schedule    
    
# i use the input function to allow for different choices in changing the classes
# as well as make the program more user friendly
    
    def add_class(self):
        
        input1 = input("Input course title (4 digit dept. code followed by 3 digit class number): ")
        input2 = input("Input current letter grade/Na if unknown: ")
        self.current_classes.update({input1: input2})
        print(f"Web-add {input1} confirmed.")
        
# this method allows to drop a class from the student's current schedule    
    
    def drop_class(self):
        
        print(self.current_classes.items())
        input3 = input("Select a course title to drop (as it appears): ")
        self.current_classes.pop(input3)
        print(f"Web-drop {input3} confirmed.")
        
# this method will return all of the students current grades in all of their classes
    
    def current_grades(self):
        
        return self.current_classes.values()
    
# the student's current classes (current_classes atribute)
# will be created first as a dictionary of key-value pairs
# as if it came from a separate database and is now being implemented into this program
    
john_grades = {"HARP 150": "A", "MATH 329": "A", "BIO 113": "B", "PSYC 111": "B"}
student1 = StudentRecord("Smith", "John", 1, "Junior", 3.5, "Biology", 2024, "NY", john_grades)

print(student1.overview())

student1.add_class()
student1.current_classes

student1.drop_class()
student1.current_classes

student1.current_grades()

```

```python
# Problem 5

# i used sets and dictionaries in my program to tell the user the following info:
# if their guess for a letter was correct, how many of that letter are in the word/phrase,
# when they have guessed all of the letters.

# i also tried using a list to print out an updated letter line as they guessed correctly
# this works perfectly fine with words that have no repeating letters,
# however for ones that do, it will only fill in the last occurence of the letter
# and not the ones tha came before.
# i believe this could have been corrected using arrays instead??? or some other method.

def hangman(word):

    print("Let's play a game! 7 guesses to spell the word, YOU MUST GUESS FOR SPACES AS WELL. Good luck!")

    print(f"The word/phrase has {len(word)} letters.")
    
    chars = set()
    for i in word:
        chars.add(i)

    char_count = {}
    for i in word:
        if i not in char_count:
            char_count[i] = 1
        else:
            char_count[i] +=1

    
    limit = 0
    letters = ["_"]*len(word)
    
    while limit < 7:
        guess = input("Guess a letter: ")
        if guess in chars:
            print("Good guess!")
            print(f"There are {char_count[guess]} {guess}'s in the word.")
            chars.remove(guess)
            index = word.rfind(guess)
            letters[index] = guess
            print(''.join(letters))
            if len(chars) == 0:
                print("You've guessed all of the letters!")
                word_guess = input("What is the word? ")
                if word_guess == word:
                    print("Correct! You Win!")
                    limit = 7
                else:
                    word_guess = input("Wrong answer.")
                    limit = 7

        else:
            print("Incorrect guess.")
            limit += 1
            print(f"You have {7-limit} guess(es) remaining.")


    print(f"The word was {word}.")
    print("Game over, thanks for playing!")


hangman("melon")
```
