```python
import requests
import json
from tkinter import *
import random

website = "https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy"

trivia = requests.get(website).json()

questions = []

for i in trivia:
    
    question = i['question']
    
    questions.append(question)
    

root = Tk()
root.geometry("900x500")

answer_choice = StringVar()
answer_choice.set("")

question_label = Label(root, text = "")

result = Label(root, text = "")

points = 0

n = len(questions)

total_points = n*10

questions_right = 0


def get_question(questions):
    
    global question
    question = random.choice(questions)
    questions.remove(question)
    
    question_label["text"] = question
    question_label.pack()
    
    
    return question
    
def get_answers(question):
    
    
    global all_answers
    all_answers = []
    
    for i in trivia:
        
        if i['question'] == question:
            
            correct_a = i['correctAnswer']
            all_answers = i['incorrectAnswers']
            all_answers.append(correct_a)
            
        else:
            pass
        
        random.shuffle(all_answers)
        
        
    return all_answers
    
def create_buttons(all_answers):
        
    global button_list
    
    button_list = []
    
    for i in all_answers:
        
        option = Radiobutton(root, text = i, value = i, command = check_answer, variable = answer_choice)
        option.pack()
        
        button_list.append(option)
        
def check_answer():
    
    answer = answer_choice.get()
    
    for i in button_list:
        
        i.destroy()
    
    for i in trivia:
        
        if i['question'] == question:
            
            if i['correctAnswer'] == answer:
                
                result["text"] = "Correct!"
                result.pack()
                
                global points
                
                points += 10
                
                global questions_right
                
                questions_right += 1
                
                
            else:
                
                result["text"] = "Incorrect :("
                result.pack()
                
            question_label["text"] = ""
            
            global button_next
            
            button_next = Button(root, text = "Next", command = restart)
            button_next.pack()
            
                
def restart():
    
    button_next.destroy()
    
    result["text"] = ""
    
    
    if len(questions) > 0:
                    
        main()
                    
    else:
        over = Label(root, text = "Game Over").pack()
        
        result["text"] = f"You got {questions_right}/{n} questions correct and scored {points}/{total_points}."
        
        result.pack()
        
        
                

def main():
    
    answer_choice = StringVar()
    answer_choice.set("")

    question_label = Label(root, text = "")

    result = Label(root, text = "")
    
    get_question(questions)
    
    get_answers(question)
    
    buttons = create_buttons(all_answers)
    

    
    
main()

root.mainloop()

# deleting radiobuttons created by for loop - source:

# https://www.daniweb.com/programming/software-development/threads/344478/tkinter-can-t-destroy-radio-buttons-created-with-a-for-loop
```
